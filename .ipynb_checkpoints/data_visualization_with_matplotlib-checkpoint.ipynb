```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(51)
y = np.arange(0,101,2)
z = x ** 3
```

#### Part - 1

        1. Figure Oluşturun.
        2. Oluşturduğunuz Figure'ün hepsini kaplayacak bir axes ekleyin.
        3. Oluturduğunuz axes üzerinde x ve y grafiği oluşturun.

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,color="red")
axes.set_xlabel("X Values")
axes.set_ylabel("Y Values")
axes.set_title("X vs Y Graph")

fig.savefig("Figure1.png")
```

#### Part - 2
        1. Figure Size'ı (8,6) olan bir figure Oluşturun
        2. axes1 ve axes2 adında 2 adet axes oluşturun.(axes1 pozisyonu = [0,0.1,0.8,0.8], axes2          pozisyonu = [0.1,0.5,0.4,0.3])
        3. axes1 üzerinde x vs z grafiğini, axes2 üzerinde x vs y grafiğini oluşturun. 

```python
fig2=plt.figure(figsize=(8,6))

axes1=fig2.add_axes([0,0.1,0.8,0.8])
axes1.plot(x,z,"purple")
axes1.set_title("Outer Graph - X vs Z Graph")

axes2=fig2.add_axes([0.1,0.5,0.4,0.3])
axes2.plot(x,y,"orange")
axes2.set_title("Inner Graph - X vs Y Graph")

fig2.savefig("Figure2.png")



```

#### Part - 3

    1. 2 adet yanyana subplot oluşturun.
    2. Birinci grafiğe (x,y) grafiğini , ikinci grafiğe (x,z) grafiğini çizin.
    3. plot() fonksiyonunun parametrelerini vererek, özelleştirin.

```python
axes=plt.subplot(1,2,1)
axes.plot(x,y,color="black",linewidth=10,linestyle="--",marker="*",markerfacecolor="yellow",markersize=10,markeredgecolor="red",label="X vs Y Graph")
axes.set_xlabel("X Values")
axes.set_ylabel("Y Values")
axes.set_title("X vs Y Graph")
axes.legend()

axes2=plt.subplot(1,2,2)
axes2.plot(x,z,color="purple",linestyle="-.",marker="o",markersize=4,markerfacecolor="black",markeredgecolor="green",label="X vs Z Graph")
axes2.set_xlabel("X Values")
axes2.set_ylabel("Z Values")
axes2.set_title("X vs Z Graph")
axes2.legend()
plt.tight_layout()

plt.savefig("Figure3.png")
```

#### Part - 4

    1. 4 adet yanyana subplot oluşturun. (figsize=(10,6))
    2. Birinci grafiğe (x,y) grafiğini , ikinci grafiğe (x,z) grafiğini çizin.
    3. Üçüncü grafiğe (x,x**2) grafiğini , dördüncü grafiğe (x,x**(0.5) grafiğini çizin.
    4. plot() fonksiyonunun parametrelerini vererek, özelleştirin.

```python
fig2,axes=plt.subplots(2,2,figsize=(10,6))

axes[0][0].plot(x,y,"red",marker="o",markerfacecolor="black")
axes[0][0].set_xlabel("X Values")
axes[0][0].set_ylabel("Y Values")
axes[0][0].set_title("X vs Y Graph")

axes[0][1].plot(x,z,"purple",marker="s",markerfacecolor="grey")
axes[0][1].set_xlabel("X Values")
axes[0][1].set_ylabel("Z Values")
axes[0][1].set_title("X vs Z Graph")

axes[1][0].plot(x,x**2,"black",marker="*",markersize=10,markerfacecolor="pink")
axes[1][0].set_xlabel("X Values")
axes[1][0].set_ylabel("X Square Values")
axes[1][0].set_title("X vs X**2 Graph")

axes[1][1].plot(x,x**(0.5),"purple",linewidth=5,linestyle="-.")
axes[1][1].set_xlabel("X Values")
axes[1][1].set_ylabel("X Square Root Values")
axes[1][1].set_title("X vs X**(0.5) Graph")

plt.tight_layout()
fig2.savefig("Figure4.png")
```

#### Part - 5

    1. 1 adet subplots oluşturun. (figsize=(8,6))
    2. Grafiğe (y=x**(0.5)) fonksiyonunu , grafiğe (y=x) fonksiyonunu çizin.
    3. Grafiğe (y=2x) fonksiyonunu , grafiğe (y=x**2) fonksiyonunu çizin.
    4. plot() fonksiyonunun parametrelerini vererek, özelleştirin.irin.

```python
x=np.arange(1,6)
fig,axes=plt.subplots(1,1,figsize=(8,6))

axes.plot(x,x**(0.5),color="purple",marker="*",label="F(x)=x**(0.5) Function")
axes.plot(x,x,color="black",marker="o",label="F(x)=x Function")
axes.plot(x,2*x,color="orange",marker="s",label="F(x)=2.x Function")
axes.plot(x,x**2,color="green",label="F(x)=x**2 Function")
axes.set_xlabel("Y Values")
axes.set_ylabel("X Values")
axes.set_title("X vs Y Funtions Graph")

plt.legend()
fig.savefig("Figure5.png")
```
