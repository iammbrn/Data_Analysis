##### Her bir videoda kullanılan etiket sayısını hesaplayarak, oluşan değerleri yeni bir sütun olarak ekleyin / Calculate the number of tags used in each video and add the resulting values ​​as a new column..

Not: Her bir etiket "|" işaretiyle ayrılmaktadır / Note: Each tag "|" It is separated by the sign..

```python
def tagsCountFind(tags):
    return len(tags.split("|"))

data["tag_count"]=data["tags"].apply(tagsCountFind)
data
```

##### (Challenge) Videoları Beğeni Oranı en yüksek olandan en düşüğe doğru sıralayınız / (Challenge) Sort the videos from the highest to the lowest Like Rate.e

```python
def likeRatioFind(count):
    return count    
data["likes_dislikesRatio"]=data["likes"].apply(likeRatioFind)/(data["likes"].apply(likeRatioFind)+data["dislikes"].apply(likeRatioFind))
data.sort_values("likes_dislikesRatio",inplace=True,ascending=False)
```

```python
def likes_Dislikes_Ratio(likes,dislikes):
    likesList=[]
    for key, value in list(data["likes"].items()):
        likesList.append(value)
    
    dislikesList=[]
    for key, value in list(data["dislikes"].items()):
        dislikesList.append(value)

    likeDislikes=list(zip(likesList,dislikesList))
    ratio=list()
    for like,dislike in likeDislikes:
        if(like+dislike==0):
            ratio.append(0)
        else:
            ratio.append(like/(like+dislike))

    return ratio

data["likes_dislikesRatio2"]=likes_Dislikes_Ratio(data["likes"],data["dislikes"])
data.sort_values("likes_dislikesRatio",inplace=True,ascending=False)

data
```
