### Region column versine göre Total score verilerinin ortalamasının büyükten küçüğe sıralamasını görüntülemek.

```python
ratio_Score=data.groupby("Region")[["Total score"]].mean() #Bölgere göre puan ortalaması bulma.
ratio_Score.sort_values("Total score",ascending=False) #Bölgelere göre puan ortalaması sıralması.
```

### Region column versi Americas olan Total score data verilerini büyükten küçüğe sıralamasını görüntülemek.

```python
data[data["Region"]=="Americas"].sort_values("Total score", ascending=False) 
```

### Region column versi Europe olan Total score data verilerini büyükten küçüğe sıralamasını görüntülemek.

```python
data[data["Region"]=="Europe"].sort_values("Total score", ascending=False).T
```

### Political regime verilerine göre Total score verilerinin ortalamasını görüntülemek.

```python
data.groupby("Political regime")[["Total score"]].mean()
```

### Political regime verilerine göre Total score verilerinin max değerlerini görüntülemek.

```python
data.groupby("Political regime").max()[["Total score"]].T
```

### Political regime verilerine göre Total score verilerinin min değerlerini görüntülemek.

```python
data.groupby("Political regime").min()[["Total score"]]
```

### Region verilerine göre Total score, Talent, Research, Infrastructure ve Operating Environment verilerinin ortalamasının Transpozunu görüntülemek.

```python
data.groupby("Region")[["Total score","Talent","Research","Infrastructure","Operating Environment"]].mean().T
```
